```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mov_time = np.genfromtxt('vmovntpd.csv', delimiter=',')
mov_time_percentage = mov_time / mov_time.sum()
```

```python
x = np.arange(1, 31)

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, mov_time_percentage)

ax.set_xlabel("Instruction number")
ax.set_ylabel("Percentage duration to total storing time")

ax.grid()

fig.savefig("vmovntpd.pdf", bbox_inches='tight')
plt.show()
```
