```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("report_resultN123MB.csv", delimiter=',', usecols=range(1,10))

rows = {"quad": 0, "hex": 6, "tet": 12, "tri": 18, "avg": 24}
figsize=(4, 3.6)
```

```python
# plot fun

def plot_function(mat_type):
    fig, ax = plt.subplots(figsize=figsize)
    row = rows[mat_type]

    ax.plot(data[row,:], data[row + 1,:], ".-", label=r"$n_B = 1$")
    ax.plot(data[row,:], data[row + 2,:], ".-", label=r"$n_B = 2$")
    ax.plot(data[row,:], data[row + 3,:], ".-", label=r"$n_B = 3$")

    ax.legend()

    plt.xticks(data[row,:])
    ax.set_xlabel(r"M blocking factor, $m_B$")
    ax.set_ylabel("Relative performance to ref. LIBXSMM")

    plt.tight_layout()
    plt.savefig(f"rt-bct-A-mem-N123B-{mat_type}.pdf", bbox_inches='tight')
    plt.show()
```

```python
plot_function("quad")
plot_function("hex")
plot_function("tet")
plot_function("tri")
plot_function("avg")
```
