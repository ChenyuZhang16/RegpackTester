```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
NB_data = np.loadtxt("report_result_NB.csv", delimiter=',', usecols=range(1,4))
MB_data = np.loadtxt("report_result_MB.csv", delimiter=',', usecols=range(1,7))
```

```python
figsize=(6, 3)
```

```python
# plot NB

fig, ax = plt.subplots(figsize=figsize)
ax.plot(NB_data[0,:], NB_data[1,:], ".-", label="hexahedral")
ax.plot(NB_data[0,:], NB_data[2,:], ".-", label="quadrilateral")
ax.plot(NB_data[0,:], NB_data[3,:], ".-", label="tetrahedral")
ax.plot(NB_data[0,:], NB_data[4,:], ".-", label="triangular")

ax.legend()
plt.xticks(NB_data[0,:])

ax.set_xlabel(r"N blocking factor, $n_B$")
ax.set_ylabel("Relative performance to ref. LIBXSMM")

plt.tight_layout()
plt.savefig("rt-bct-A-mem-NB.pdf", bbox_inches='tight')
plt.show()
```

```python
# plot MB

fig, ax = plt.subplots(figsize=figsize)
ax.plot(MB_data[0,:], MB_data[1,:], ".-", label="hexahedral")
ax.plot(MB_data[0,:], MB_data[2,:], ".-", label="quadrilateral")
ax.plot(MB_data[0,:], MB_data[3,:], ".-", label="tetrahedral")
ax.plot(MB_data[0,:], MB_data[4,:], ".-", label="triangular")

ax.legend()
plt.xticks(MB_data[0,:])

ax.set_xlabel(r"M blocking factor, $m_B$")
ax.set_ylabel("Relative performance to ref. LIBXSMM")

plt.tight_layout()
plt.savefig("rt-bct-A-mem-MB.pdf", bbox_inches='tight')
plt.show()
```
